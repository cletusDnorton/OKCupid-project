```python
cols =['body_type', 'diet', 'orientation', 'education', 'religion_cleaned',
       'sex', 'job', 'sign_cleaned']

profile_df = df[cols].dropna()
profile_df.shape
```

```python
for column in cols[:-1]:
    profile_df = pd.get_dummies(profile_df, columns=[column], prefix = [column])
```

```python
profile_df.head()
```

```python
df.sign_cleaned.value_counts()
```

```python
col_length = len(profile_df.columns)
X = profile_df.iloc[:, 1:col_length]
Y = profile_df.iloc[:, 0:1]

from sklearn.model_selection import train_test_split 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state = 0)

Y_train = Y_train.to_numpy().ravel()
Y_test = Y_test.to_numpy().ravel()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors = 5)
knn_model.fit(X_train, Y_train)
knn_predictions = knn_model.predict(X_train)
```

```python
print('Mean accuracy is: ' + str(knn_model.score(X_train,Y_train)))
```

```python
from sklearn.metrics import precision_recall_fscore_support
precision_recall_fscore_support(Y_train, knn_predictions)
```

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(knn_model, X_test, Y_test, xticks_rotation='vertical')
plt.show()
plt.savefig('Confusion matrix KNN')
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_class = DecisionTreeClassifier()
tree_class.fit(X_train, Y_train)
```

```python
print(tree_class.score(X_train, Y_train))
prediction = tree_class.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

label = tree_class.classes_
print(label)
matrix = confusion_matrix(Y_test, prediction)
```

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(tree_class, X_test, Y_test, xticks_rotation='vertical')
plt.show()
plt.savefig('Confusion Matrix Decision Tree')
```

```python
tree_class.get_depth()
```

```python
tree_class20 = DecisionTreeClassifier(max_depth = 20)
tree_class20.fit(X_train, Y_train) 
tree_predictions20 = tree_class20.predict(X_train)
tree_class20.score(X_train, Y_train)
```

```python
plot_confusion_matrix(tree_class20, X_test, Y_test, xticks_rotation='vertical')
plt.show()
plt.savefig('Confusion Matrix Decision Tree limit depth')
```
